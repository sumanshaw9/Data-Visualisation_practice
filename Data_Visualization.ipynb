**3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.**

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
df
```

```python
aggregated_data = df.groupby(['Month', 'Day']).agg({'Sales': 'mean'}).reset_index()

# Create a pivot table
pivot_table = aggregated_data.pivot(index='Day', columns='Month', values='Sales')

# Create the heatmap using Plotly Express
fig = px.imshow(
    pivot_table,
    color_continuous_scale='Viridis',
    labels=dict(x="Month", y="Day", color="Average Sales")
)

# Show the plot
fig.show()
```

**5. Create a Bokeh heatmap using the provided dataset.**


```python
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.palettes import Viridis256
from bokeh.transform import transform
  #given
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
# Flatten the meshgrid and data for use with Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=values))

mapper = LinearColorMapper(palette=Viridis256, low=values.min(), high=values.max())

p = figure(title="Heatmap Example", x_axis_label='X', y_axis_label='Y',
           toolbar_location=None,)
p.rect(x='x',y='y',width=0.1, height=0.1,source=source,line_color=None, fill_color=transform('value', mapper))

# Add color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
show(p)
```
